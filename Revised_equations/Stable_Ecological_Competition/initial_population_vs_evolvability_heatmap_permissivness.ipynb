```python
import numpy as np
import plotly.graph_objects as go
import plotly.subplots as sp


# Function to generate parameters with varied permissiveness
def generate_parameters_with_varied_parameters(N2_N1_ratios, V_A2_V_A1_ratios, permissiveness_values):
    parameters_list = []
    for N2_N1 in N2_N1_ratios:
        for V_A2_V_A1 in V_A2_V_A1_ratios:
            for permissiveness in permissiveness_values:
                parameters_dict = Parameters_for_Analysis(base_parameters=base_parameters, V_A1=0.01, V_A2=0.01).__dict__.copy()
                parameters_dict['initial_population_species_2'] = parameters_dict['initial_population_species_1'] * N2_N1
                parameters_dict['V_A2'] = parameters_dict['V_A1'] * V_A2_V_A1
                parameters_dict['mean_heterospecifc_permissivness_females_sp_1'] = permissiveness
                parameters_dict['mean_heterospecifc_permissivness_females_sp_2'] = permissiveness
                parameters = Parameters_for_Analysis(base_parameters=base_parameters, **parameters_dict)
                parameters_list.append(parameters)

    return parameters_list

# Function to run sensitivity analysis and return a 3D tensor of outcomes
def run_sensitivity_analysis(parameters_list, N2_N1_ratios, V_A2_V_A1_ratios, permissiveness_values):
    final_outcomes_tensor = np.zeros((len(N2_N1_ratios), len(V_A2_V_A1_ratios), len(permissiveness_values)), dtype=object)
    
    index = 0
    for i, N2_N1 in enumerate(N2_N1_ratios):
        for j, V_A2_V_A1 in enumerate(V_A2_V_A1_ratios):
            for k, permissiveness in enumerate(permissiveness_values):
                parameters = parameters_list[index]
                final_outcome_instance = run_simulation_and_get_outcome_instance(parameters)
                final_outcomes_tensor[i, j, k] = final_outcome_instance.outcome
                index += 1

    return final_outcomes_tensor

# Function to plot 2D sensitivity analysis results as heatmaps
def plot_3d_sensitivity_analysis_initial_pop_vs_evolvability(N2_N1_ratios, V_A2_V_A1_ratios, final_outcomes_tensor, permissiveness_values):
    color_map = {
        "sp 2 (high-genetic-var) wins, sp 1 (low-genetic-var) extinct": 0,
        "sp 1 (low-genetic-var) wins, sp 2 (high-genetic-var) extinct": 1,
        "both coexist": 2,
        "both extinct": 3,

    }
    
    # Select 4 permissiveness values to plot
    selected_permissiveness_indices = [0,1,2,3,4,5,6,7,8]   # Adjust these indices if needed

    # Create a subplot grid (2x2 for the 4 heatmaps)
    fig = sp.make_subplots(rows=3, cols=3, subplot_titles=[f'Permissiveness = {permissiveness_values[k]:.2f}' for k in selected_permissiveness_indices])

    # Iterate over the subplots
    for idx, k in enumerate(selected_permissiveness_indices):
        # Get the 2D slice for the current permissiveness value
        outcomes_at_permissiveness = final_outcomes_tensor[:, :, k]

        # Convert outcome strings to color indices using the color_map
        color_matrix = np.array([[color_map[outcomes_at_permissiveness[i, j]] for j in range(len(V_A2_V_A1_ratios))] for i in range(len(N2_N1_ratios))])

        # Create a heatmap for this slice
        heatmap = go.Heatmap(
            z=color_matrix,
            x=np.log10(V_A2_V_A1_ratios),  # Log scale for V_A2/V_A1
            y=np.log10(N2_N1_ratios),      # Log scale for N2/N1
             colorscale=[
                [0.0, 'red'],
                [0.25, 'red'],
                [0.25, 'blue'],
                [0.5, 'blue'],
                [0.5, 'green'],
                [0.75, 'green'],
                [0.75, 'black'],
                [1.0, 'black']
            ],
            zmin=0,
            zmax=3,
            showscale=False,
            hoverinfo='none'
        )

        # Add the heatmap to the subplot
        row, col = divmod(idx, 3)
        fig.add_trace(heatmap, row=row+1, col=col+1)

    # Update layout for axis labels and titles
    fig.update_layout(
        height=800, width=800,
        title="Heatmaps for Different Permissiveness Values",
        xaxis_title="log(V_A2/V_A1 Ratios)",
        yaxis_title="log(N2/N1 Ratios)"
    )

    # Set common axis labels for all subplots
    for i in range(1, 10):
        fig['layout'][f'xaxis{i}'].update(title='log(V_A2/V_A1 Ratios)')
        fig['layout'][f'yaxis{i}'].update(title='log(N2/N1 Ratios)')

    fig.show()

# Main function to run the analysis
def main():
    N2_N1_powers = np.arange(-3, 3, 0.0625)
    V_A2_V_A1_powers = np.arange(0, 3, 0.0625)
    
    N2_N1_ratios = np.power(10.0, N2_N1_powers)
    V_A2_V_A1_ratios = np.power(10.0, V_A2_V_A1_powers)
    permissiveness_values = np.linspace(0, 1, 10)
    
    parameters_list = generate_parameters_with_varied_parameters(N2_N1_ratios, V_A2_V_A1_ratios, permissiveness_values)
    final_outcomes_tensor = run_sensitivity_analysis(parameters_list, N2_N1_ratios, V_A2_V_A1_ratios, permissiveness_values)
    
    plot_3d_sensitivity_analysis_initial_pop_vs_evolvability(N2_N1_ratios, V_A2_V_A1_ratios, final_outcomes_tensor, permissiveness_values)

if __name__ == "__main__":
    main()


```
