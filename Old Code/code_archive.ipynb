```python
def mating_compatibility(male_type, female_type):
    if male_type == "A":
        if female_type == "A":
            mating_compatibility = A_males_preference_for_A_female*A_females_preference_for_A_male

        if female_type == "B":
            mating_compatibility = A_males_preference_for_B_female*B_females_preference_for_A_male
        
    if male_type == "B":

        if female_type == "A":
            mating_compatibility = B_males_preference_for_A_female*A_females_preference_for_B_male

        if female_type == "B":
            mating_compatibility = B_males_preference_for_B_female*B_females_preference_for_B_male

    return mating_compatibility
```

class Simulation:
    def __init__(self, params):
        self.params = params
        self.population_species_1 = np.zeros(params.num_generations)
        self.population_species_2 = np.zeros(params.num_generations)
        self.population_species_1[0] = params.initial_population_species_1

    def next_generation_ecology(self, t):
        N1 = self.population_species_1[t]
        N2 = self.population_species_2[t]
        
        next_N1 = self.params.b1 * (N1 / 2) * (N1 / (N1 + self.params.strength_of_interference * (1 - self.z_m2) * N2)) - self.params.k1 * N1 * (N1 + self.params.c12 * N2)
        next_N2 = self.params.b2 * (N2 / 2) * (N2 / (N2 + self.params.strength_of_interference * (1 - self.z_m1) * N1)) - self.params.k2 * N2 * (N2 + self.params.c21 * N1)
        
        self.population_species_1[t + 1] = max(next_N1, 0)
        self.population_species_2[t + 1] = max(next_N2, 0)

    def next_generation_traits(self, N1, N2):
        W_m1 = 1 - (self.params.cost_to_males * (1 - self.z_m1) * (N2 / (N2 + N1)))
        W_m2 = 1 - (self.params.cost_to_males * (1 - self.z_m1) * (N2 / (N2 + N1)))
        
        dW_m1_dz_m2 = self.params.cost_to_males * (N2 / (N2 + N1))
        dW_m2_dz_m2 = self.params.cost_to_males * (N1 / (N1 + N2))
        
        delta_z_m1 = self.params.V_A1 * (1 / W_m1) * dW_m1_dz_m2
        delta_z_m2 = self.params.V_A2 * (1 / W_m2) * dW_m2_dz_m2
        
        new_z_m1 = self.z_m1 + delta_z_m1
        new_z_m2 = self.z_m2 + delta_z_m2
        
        if new_z_m1 <= 1:
            self.z_m1 = new_z_m1
        if new_z_m2 <= 1:
            self.z_m2 = new_z_m2
        
        self.z_m1_history.append(self.z_m1)
        self.z_m2_history.append(self.z_m2)

        N1 = self.population_species_1[t + 1]  # Updated populations for the next generation

def plot_2d_sensitivity_analysis(x_ratios, y_ratios, final_outcomes_matrix, x_label, y_label, x_tickvals, x_ticktext, y_tickvals, y_ticktext):
    color_map = {
        "sp 2 (high-genetic-var) wins, sp 1 (low-genetic-var) extinct": 0,
        "sp 1 (low-genetic-var) wins, sp 2 (high-genetic-var) extinct": 1,
        "both extinct": 2,
        "both coexist": 3
    }

    # Convert final outcomes to numerical values
    numerical_matrix = np.vectorize(color_map.get)(final_outcomes_matrix)

    # Create the heatmap
    heatmap = go.Heatmap(
        z=numerical_matrix.T,  # Transpose the matrix to match axis
        x=x_ratios,
        y=y_ratios,
        colorscale=[[0, 'red'], [1/3, 'blue'], [2/3, 'black'], [1, 'green']],
        showscale=False,
        colorbar=dict(
            tickvals=[0, 1, 2, 3],
            ticktext=list(color_map.keys()),
        )
    )

    fig = go.Figure(data=[heatmap])

    # Adding legend for outcomes
    legend_labels = [
        ("Slow evolver extinct", 'red'),
        ("Fast evolver extinct", 'blue'),
        ('Both coexist', 'green'),
        ('Both extinct', 'black')
    ]

    for label, color in legend_labels:
        fig.add_trace(go.Scatter(
            x=[None], y=[None], mode='markers',
            marker=dict(color=color),
            showlegend=True,
            name=label
        ))

    # Update plot layout with provided labels and tick values
    fig.update_layout(
        xaxis=dict(
            title=x_label,
            type='log',
            tickvals=x_tickvals,
            ticktext=x_ticktext
        ),
        yaxis=dict(
            title=y_label,
            type='log',
            tickvals=y_tickvals,
            ticktext=y_ticktext
        ),
        template='plotly',
        width=1000,
        height=800,
    )

    pio.show(fig)

# Function to generate parameters based on any x/y axis variable
def generate_parameters_with_varied_parameters(x_ratios, y_ratios, x_param_name, y_param_name):
    parameters_list = []
    for x_ratio in x_ratios:
        for y_ratio in y_ratios:
            parameters_dict = Parameters_for_Analysis(base_parameters=base_parameters, V_A1=0.01, V_A2=0.01).__dict__.copy()

            # Dynamically assign the values based on x and y parameter names
            parameters_dict[x_param_name] = parameters_dict['initial_population_species_1'] * x_ratio
            parameters_dict[y_param_name] = parameters_dict['V_A1'] * y_ratio

            # Create a new Parameters instance with the modified dictionary
            parameters = Parameters_for_Analysis(base_parameters=base_parameters, **parameters_dict)
            parameters_list.append(parameters)
    
    return parameters_list

# Function to run sensitivity analysis
def run_sensitivity_analysis(parameters_list):
    final_outcomes_list = []
    for parameters in parameters_list:
        final_outcome_instance = run_simulation_and_get_outcome_instance(parameters)
        final_outcomes_list.append(final_outcome_instance.outcome)
    return final_outcomes_list

# Main function that accepts x and y axis variables
def main(x_param_name="initial_population_species_2", y_param_name="V_A2"):
    x_powers = np.arange(-3, 3, 0.0625)
    y_powers = np.arange(0, 3, 0.0625)
    
    # Dynamically set x and y ratios
    x_ratios = np.power(10.0, x_powers)
    y_ratios = np.power(10.0, y_powers)
    
    # Generate parameters based on the chosen x and y variables
    parameters_list = generate_parameters_with_varied_parameters(x_ratios, y_ratios, x_param_name, y_param_name)
    
    # Run the sensitivity analysis
    final_outcomes_list = run_sensitivity_analysis(parameters_list)
    
    # Reshape results into a matrix
    final_outcomes_matrix = np.array(final_outcomes_list).reshape(len(x_ratios), len(y_ratios))
    
    # Define x and y tick values for log scales
    x_tickvals = [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]
    x_ticktext = ["0.001", "0.01", "0.1", "1", "10", "100", "1000"]
    y_tickvals = [1, 10, 100, 1000]
    y_ticktext = ['1', '10', '100', '1000']
    
    # Plot the results
    plot_2d_sensitivity_analysis(
        x_ratios, y_ratios, final_outcomes_matrix,
        x_label=f'{x_param_name} (log-scale)',
        y_label=f'{y_param_name} (log-scale)',
        x_tickvals=x_tickvals,
        x_ticktext=x_ticktext,
        y_tickvals=y_tickvals,
        y_ticktext=y_ticktext
    )

if __name__ == "__main__":
